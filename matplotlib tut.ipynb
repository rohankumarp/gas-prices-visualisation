```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# #To Load The Dataframe

gas = pd.read_csv('gas_prices.csv')
plt.figure(figsize=(20,10))

# Give it a name

plt.title("Gas prices in USD")

#One Way of plotting

plt.plot(gas.Year, gas.USA,label = "USA" , color='b',marker='x',linestyle='-' )
plt.plot(gas["Year"],gas.Canada,label="Canada", color='g',marker='o',linestyle='-')
plt.plot(gas["Year"],gas["South Korea"],label="South Korea" , color='orange',marker='x',linestyle='-')

#Another Way of plotting

for country in gas:
    if country != "Year":
     plt.plot(gas.Year , gas[country] , label=country , marker = 'x' , linestyle="-")
    
    #Labe and stuff
plt.xticks(gas.Year[::3],rotation = 'vertical')
plt.xlabel("Years")
plt.ylabel("Price in USD")
plt.legend()

# to save the figure

plt.savefig("Gas_price_visual")

plt.show()
```
